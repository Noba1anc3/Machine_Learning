```python
import numpy as np
```

```python
arr1 = np.arange(2,14)
print(arr1)
```

```python
print(arr1[2])#第二个位置的数据
```

```python
print(arr1[1:4])#第一到第四个位置的数据
```

```python
print(arr1[2:-1])#第二到倒数第一个位置的数据
```

```python
print(arr1[:5])#前五个数据
```

```python
print(arr1[-2:])#最后两个数据
```

```python
arr2 = arr1.reshape(3,4)
print(arr2)
```

```python
print(arr2[1])
```

```python
print(arr2[1][1])
```

```python
print(arr2[1,2])
```

```python
print(arr2[:,2])
```

```python
for i in arr2: #迭代行
    print(i)
```

```python
for i in arr2.T:#迭代列
    print(i)
```

```python
for i in arr2.flat:#一个一个元素迭代
    print(i)
```
