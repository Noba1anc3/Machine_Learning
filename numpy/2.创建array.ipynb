```python
import numpy as np
```

```python
a = np.array([1,2,3],dtype=np.int32)
print(a.dtype)
```

```python
b = np.array([1,2,3],dtype=np.float)
print(b.dtype)
```

```python
c = np.array([1,2,3])#一维数组
print(c)
```

```python
d = np.array([[1,2,3],   #2维矩阵
                      [4,5,6]])
print(d)
```

```python
zero = np.zeros((2,3)) #生成2行3列全为0的矩阵
print(zero)
```

```python
one = np.ones((3,4)) #生成3行4列全为1的矩阵
print(one)
```

```python
empty = np.empty((3,2))#生成3行2列全都接近于0（不等于0）的矩阵
print(empty)
```

```python
e = np.arange(10)
print(e)
```

```python
f = np.arange(4,12)
print(f)
```

```python
g = np.arange(1,20,3)
print(g)
```

```python
np.arange(8)
```

```python
h = np.arange(8).reshape(4,2)#重新定义矩阵的形状
print(h)
```
