```python
import numpy as np
```

```python
arr1 = np.array([[1,2,3],
                           [4,5,6]])
arr2 = np.array([[1,1,2],
                           [2,3,3]])
print(arr1)
print(arr2)
```

```python
print(arr1 + arr2)
```

```python
print(arr1 - arr2)
```

```python
print(arr1 * arr2)
```

```python
print(arr1 ** arr2)
```

```python
print(arr1 / arr2)
```

```python
print(arr1 % arr2)
```

```python
print(arr1 // arr2)
```

```python
print(arr1+2)#所有的元素加2
```

```python
print(arr1*10)#所有的元素乘以10
```

```python
arr3 = arr1 > 3 #判断哪些元素大于3
print(arr3)
```

```python
arr4 = np.ones((3,5))
print(arr4) 
```

```python
print(arr1)        
```

```python
np.dot(arr1,arr4)#矩阵乘法   2*3  3*5
```
