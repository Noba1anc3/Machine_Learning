```python
a_list = [1,2,30,30,30,4,2]
print(a_list)
```

```python
a_list[0] = 100 #修改列表中第0个元素
print(a_list) 
```

```python
a_list.append(200) #在列表末尾添加一个元素
print(a_list)
```

```python
a_list.insert(2,300) #在列表中插入一个元素
print(a_list)
```

```python
del a_list[2] #删除列表第2个元素
print(a_list)
```

```python
a_list.remove(30) #删除列表中的一个‘30’
print(a_list)
```

```python
a = a_list.pop()
print(a)
print(a_list)
```
